# Predicting Car Prices

```python
# importing packages and read data

import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
%matplotlib inline

cars = pd.read_csv('imports-85.data')

header = ['symboling','normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
cars.columns = header

cars.head(10)
```

## Checking and Cleaning Data

```python
# Replacing '?' in the data with nan then see which column has nan

print(cars.info())
cars = cars.replace('?',np.nan)
print(cars.isnull().sum())
```

```python
# clear out na
cars = cars.dropna(subset=['price'])
print(cars.isnull().sum())

# Replacing string to number so we can change to float/int
num = {'eight':8, 'five':5, 'four':4, 'six':6, 'three':3, 'twelve':12, 'two':2}
cars=cars.replace({'num-of-doors':num,'num-of-cylinders':num})
cars = cars.astype({'normalized-losses':float,'bore':float,'stroke': float,'horsepower':float,'peak-rpm':float,'price':int})

# Averaging out missing values in columns
cars = cars.fillna(cars.mean())
#cars['normalized-losses'] = cars['normalized-losses'].fillna(cars['normalized-losses'].mean())

print(cars.info())
print(cars.head(10))
print(cars.isnull().sum())
```

```python
# excluding non numeric columns and convert to float
numeric_cars = cars.select_dtypes(exclude=['object']) #!= object
columns_numeric = numeric_cars.columns.tolist()
price_column = numeric_cars['price']
numeric_cars = numeric_cars.astype('float')
print(price_column.shape)
print(numeric_cars.info())
print(numeric_cars.index)

# Normalize the data using Normalization formula = (x-min(x))/(max(x)-min(x))
normalized_cars = (numeric_cars - numeric_cars.min())/(numeric_cars.max() - numeric_cars.min())
print(normalized_cars.index)

#re-add non-normalized price column
normalized_cars['price'] = price_column
print(normalized_cars.info())
print(normalized_cars.isnull().sum())
print(normalized_cars.head(10))
```

## Univariate Model

```python
# Define our k nearest neighbor (knn) function
def knn_train_test(train_col,target_col,df):
    # Randomize the dataset
    np.random.seed(1)
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)

    # splitting the dataset into train and test datasets
    train_df = rand_df[:145]
    test_df = rand_df[145:]
    
    # train the knn and make predictions on test df
    knn = KNeighborsRegressor()
    knn.fit(train_df[[train_col]],train_df[target_col])
    predictions = knn.predict(test_df[[train_col]])
    
    # find root mean squared error
    rmse = mean_squared_error(test_df[target_col],predictions) ** (1/2)
    return rmse
    
```

```python
train_cols = normalized_cars.drop(['price'], axis = 1).columns
rmses = {}
lowest_rmse ={}
for train_col in train_cols:
    rmse = knn_train_test(train_col,'price',normalized_cars)
    rmses[train_col] = rmse

print(rmses, '\n')

# Convert rmses to series for better display
rmses_series = pd.Series(rmses)
print(rmses_series.sort_values(),'\n')

# Get lowest rmse
min_key = min(rmses, key = rmses.get)
lowest_rmse[min_key] = rmses[min_key]

print(lowest_rmse)

# Visualize the results
x = [x for x in range(len(train_cols))]
plt.scatter(list(rmses.keys()),list(rmses.values()))
plt.xticks(rotation=90)
plt.show()
```

Engine size gives the lowest RMSE of 2690 for our univariate model at default k = 5

```python
# Change our function so it accepts k value
def knn_train_test_k(train_col,target_col,df,k):
    # Randomize the dataset
    np.random.seed(1)
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)

    # splitting the dataset into train and test datasets
    train_df = rand_df[:145]
    test_df = rand_df[145:]
    
    # train the knn and make predictions on test df
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(train_df[[train_col]],train_df[target_col])
    predictions = knn.predict(test_df[[train_col]])
    
    # find root mean squared error
    rmse = mean_squared_error(test_df[target_col],predictions) ** (1/2)
    return rmse
    
```

```python
# Train our model using different k values and iterate for each column
k_values = [1,3,5,7,9]

total_rmses = {}

for train_col in train_cols:
    rmses_k = {}
    for k in k_values:
        rmse_k = knn_train_test_k (train_col,'price',normalized_cars,k)
        rmses_k[k] = rmse_k
    total_rmses[train_col] = rmses_k       

# Visualize our results
for kee,val in total_rmses.items():
    x = list(val.keys())
    y = list(val.values())
    print(kee,val)
    plt.plot(x,y)
    plt.xlabel('k')
    plt.ylabel('RMSE')

# Turn our results into a dataframe for easy data manipulation
total_rmses_df = pd.DataFrame(data=total_rmses)
print(total_rmses_df)

# Minimum rmse for each column
min_rmse_each_col = pd.DataFrame({'min': total_rmses_df.min(), 'k value': total_rmses_df.idxmin()})
print('\nMin RMSE for each column:\n',min_rmse_each_col.sort_values('min'))

# Mean rmse for each colum
print('\nMean RMSE for each column:\n',total_rmses_df.mean().sort_values())
```

Looking at the results, curb weight gives the lowest RMSE of 2591 at k = 7 and on average, engine size will give lowest RMSE

## Multivariate Model

```python
# Modify our k nearest neighbor (knn) function to accept multiple columns
def multi_knn_train_test(train_col,target_col,df,k):
    
    # Randomize the dataset
    np.random.seed(1)
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)

    # splitting the dataset into train and test datasets
    train_df = rand_df[:145]
    test_df = rand_df[145:]
    
    # train the knn and make predictions on test df
    knn = KNeighborsRegressor(n_neighbors = k)
    knn.fit(train_df[train_col],train_df[target_col])
    predictions = knn.predict(test_df[train_col])
    
    # find root mean squared error
    rmse = mean_squared_error(test_df[target_col],predictions) ** (1/2)
    return rmse
```

```python
# Train model on all columns 
rmse = multi_knn_train_test(list(train_cols),'price',normalized_cars,5)
print(rmse)
```

```python
# Train model on best 2 columns 
best_two = ['curb-weight','engine-size']
rmse_best_two = multi_knn_train_test(best_two,'price',normalized_cars,5)
print('RMSE for best two features: ', rmse_best_two)

# Train model on best 3 columns 
best_three = ['curb-weight','engine-size','width']
rmse_best_three = multi_knn_train_test(best_three,'price',normalized_cars,5)
print('RMSE for best three features: ', rmse_best_three)

# Train model on best 4 columns 
best_four = ['curb-weight','engine-size','width','city-mpg']
rmse_best_four = multi_knn_train_test(best_four,'price',normalized_cars,5)
print('RMSE for best four features: ', rmse_best_four)

# Train model on best 5 columns 
best_five = ['curb-weight','engine-size','width','city-mpg','highway-mpg']
rmse_best_five = multi_knn_train_test(best_five,'price',normalized_cars,5)
print('RMSE for best five features: ', rmse_best_five)
```

```python
hyper_params = [k for k in range(1,26)]
rmses_top = {}
rmses_top_two = {}
rmses_top_three = {}
rmses_all = {}

for k in hyper_params:
    rmse_top = multi_knn_train_test(best_five,'price',normalized_cars,k)
    rmse_top_two = multi_knn_train_test(best_four,'price',normalized_cars,k)
    rmse_top_three = multi_knn_train_test(best_three,'price',normalized_cars,k)
    
    rmses_top[k] = rmse_top
    rmses_top_two[k] = rmse_top_two
    rmses_top_three[k] = rmse_top_three

rmses_all['top - best five'] = rmses_top
rmses_all['top two - best four'] = rmses_top_two
rmses_all['top three - best three'] = rmses_top_three

# Turn our results into a dataframe for easy data manipulation
rmses_all_df = pd.DataFrame(data=rmses_all)
print(rmses_all_df)

# Minimum rmse for top 1,2,3
min_rmse_top = pd.DataFrame({'min': rmses_all_df.min(), 'k value': rmses_all_df.idxmin()})
print('\nMin RMSE for top 1,2,3:\n',min_rmse_top.sort_values('min'))

# Visualize our results
for k,v in rmses_all.items():
    x = list(v.keys())
    y = list(v.values())
    plt.plot(x,y)
    plt.xlabel('k')
    plt.ylabel('RMSE')
```

Based on the results, best four features ('curb-weight','engine-size','width','city-mpg') gives the lowest RMSE of 1966 at k = 2

## KFold Cross Validation Method

```python
from sklearn.model_selection import cross_val_score

def kfold_train_test_k(train_col,target_col,df,k):
    
    model = KNeighborsRegressor(n_neighbors=k)
    mses = cross_val_score(model, df[[train_col]], df[target_col], scoring="neg_mean_squared_error", cv=10)
    rmses = np.sqrt(np.abs(mses))
    avg_rmse = np.mean(rmses)
    
    return avg_rmse

kf_all_rmses = {}
for train_col in train_cols:
    kf_rmses = {}
    for k in k_values:
        kf_rmse = kfold_train_test_k(train_col,'price',normalized_cars,k)
        kf_rmses[k] = kf_rmse
    kf_all_rmses[train_col] = kf_rmses

# Turn our results into a dataframe for easy data manipulation
kf_all_rmses_df = pd.DataFrame(data=kf_all_rmses)
#print(total_rmses_df)

# Minimum rmse for each column
kf_min_rmse_each_col = pd.DataFrame({'min': kf_all_rmses_df.min(), 'k value': kf_all_rmses_df.idxmin()})
print('\nMin RMSE for each column:\n',kf_min_rmse_each_col.sort_values('min'))

# Mean rmse for each colum
print('\nMean RMSE for each column:\n',kf_all_rmses_df.mean().sort_values())
```

## GridsearchCV

```python
from sklearn.model_selection import GridSearchCV

hyperparameters = {
    'n_neighbors': k_values,
    'algorithm': ['brute'],
    'p': [2]
}

knn = KNeighborsRegressor()
grid = GridSearchCV(knn, param_grid = hyperparameters, cv=10, scoring = 'neg_mean_squared_error')
grid.fit(normalized_cars[train_cols],normalized_cars['price'])
best_params = grid.best_params_
best_score = grid.best_score_
rmse_best_score = abs(best_score)**(1/2)

print(best_params)
print(rmse_best_score)
```
